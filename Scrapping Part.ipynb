```python
import time
import pandas as pd
from bs4 import BeautifulSoup as bs4
import requests
import html
import urllib.request
import os
```

```python
def createData():
    for filename in os.listdir("."):
        if filename.endswith(".html"): 
             ParseHtml(filename)
```

```python
def ParseHtml(fileName):
    f = open(fileName, 'r', encoding="utf-8")
    f2 = open("youtube2.csv" , 'a')
    data = f.read()
    data = bs4(data)
    data = data.findAll('a',{'class':'yt-simple-endpoint style-scope ytd-video-renderer'})
    x = 1
    for i in data:
        link = i['href']
        link = "https://www.youtube.com" + link
        re = requests.get(link)
        link  = link.split("=")[-1]
        title = i['title']
        htmltext = re.text
        htmltxt = bs4(htmltext)
        try:
            desc = htmltxt.findAll('p',{'id':'eow-description'})[0].text
            f2.write(link+','+str(title.replace("," , " "))+','+str(desc.replace("," , " "))+','+fileName.split(".")[0]+'\n')
            print(fileName.split(".")[0] + str(x) + "/"+ str(len(data)))
            x = x+1
        except:
            pass
        
        
    
```

```python
createData()
```
